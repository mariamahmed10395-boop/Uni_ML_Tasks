```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

df = sns.load_dataset('titanic')
```

```python
df
```

```python
df.info()
```

```python
df_numeric = df.select_dtypes(include=[np.number])
```

```python
total_missing = df_numeric.isnull().sum().sum()
total_cells = np.prod(df_numeric.shape)
total_percent = (total_missing / total_cells) * 100
```

```python
total_missing
```

```python
total_percent
```

```python
col_missing_percent = (df_numeric.isnull().sum() / len(df_numeric)) * 100
print("\nMissing percentage per column:")
print(col_missing_percent)
```

```python
imputer = IterativeImputer(max_iter=10, random_state=42)
```

```python
df_imputed_array = imputer.fit_transform(df_numeric)
```

```python
df_final = pd.DataFrame(df_imputed_array, columns=df_numeric.columns)
```
