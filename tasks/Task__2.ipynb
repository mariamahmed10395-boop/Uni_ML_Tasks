```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data = {
    'Country': ['France', 'Spain', 'Germany', 'Spain', 'Germany'],
    'Age': [44, 27, 30, 38, 40],
    'Salary': [72000, 48000, 54000, 61000, 50000],
    'Purchased': ['No', 'Yes', 'No', 'No', 'Yes']
}
df = pd.DataFrame(data)

```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Purchased'] = label_encoder.fit_transform(df['Purchased'])

print(df)
```

```python

encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(df[['Country']])

encoded_df = pd.DataFrame(
    one_hot_encoded, 
    columns=encoder.get_feature_names_out(['Country'])
)
```

```python
cols = list(df.columns)
target_idx = cols.index('Country')

df_temp = df.drop(columns=['Country'])

for i, col_name in enumerate(encoded_df.columns):
    df_temp.insert(target_idx + i, col_name, encoded_df[col_name])

print(df_temp)
```
